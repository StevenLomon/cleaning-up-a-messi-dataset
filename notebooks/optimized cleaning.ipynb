```python
import kagglehub
import pandas as pd
```

```python
path = kagglehub.dataset_download("yagunnersya/fifa-21-messy-raw-dataset-for-cleaning-exploring")
```

```python
df = pd.read_csv(f"{path}/fifa21 raw data v2.csv")
```

```python
df
```

```python
# Function to fix cases where LongName contains an abbreviated first name (e.g., "L. Messi")
def fix_abbreviated_name(row):
    # Check if there's a first name in 'Name' and if 'LongName' has an abbreviation
    if len(row['Name'].split()) > 1 and row['Name'][1] != '.':
        first_name = row['Name'].split()[0]
        # Replace the abbreviated first name in 'LongName' with the full 'FirstName'
        row['LongName'] = f"{first_name} {row['LongName'][2:]}"
    return row

# Apply the function only to rows that match the abbreviation condition
abbreviation_condition = df['LongName'].str[1] == '.'  # Only rows with abbreviated first names
df.loc[abbreviation_condition] = df.loc[abbreviation_condition].apply(fix_abbreviated_name, axis=1)

# Instead of dropping the Name column and inserting a new FirstName column, we can simply rename it to FirstName
df = df.rename(columns={'Name':'FirstName'})

# And now on to splitting the LongName column and renaming it LastName
df['FirstName'] = df['LongName'].apply(lambda x: x.split()[0])
df['LongName'] = df['LongName'].apply(lambda x: " ".join(x.split()[1:]))  # Remove the first name in LongName

df = df.rename(columns={'LongName':'LastName'})
```
