Okay, wow, it's not as easy as dropping the Name column and splitting the name in LongName which we can see with Mr. Portugal haha!<br>
This is gonna require some finessing. I want to believe that the only exception is the ones that Ronaldo has; that the first name is simply a letter. And in these cases, we need to grab the first name from the name column. And after this operation, *then* we can drop the Name column

```python
exceptions = df[df['LongName'].apply(lambda x: x[1] == '.' if len(x) > 1 else False)]
# exceptions = df[df['LongName'].str.contains(r'\.')] This would check simply for the presence of a dot but that doesn't give us what we want
exceptions

# Saving the IDs of the exceptions to double-check after the transformation of df
exceptions_ids = list(exceptions['ID'])
```

These are the exceptions we need to handle. In all cases except for player #256612 and #190883 we can simply grab the first name from the Name column like we initially planned. For these other two, there will be two cases; either Name.split() is of length 1 (which is the case for Evanilson), or there is no first name to be grabbed (which is the case for A. DeLaGarza). Let's start coding this up:

```python
for index, row in exceptions.iterrows():
    if len(row['Name'].split()) > 1 and row['Name'][1] != '.':
        # print(f"Name: {row['Name']}, LongName: {row['LongName']}")
        first_name = row['Name'].split()[0]
        df.at[index, 'LongName'].replace('.', '')
        df.at[index, 'LongName'] = f"{first_name} {row['LongName'][2:]}"
```

There is essentially an else: pass in the for loop above since in the two exceptions I think the best solution is to just leave it as is (F. Evanilson de Lima Barbosa and A.J. DeLaGarza)

```python
df_exceptions_checked = df[df['ID'].isin(exceptions_ids)]
df_exceptions_checked
```

```python
# *Now* we can drop the Name column
#df = df.drop(columns=['Name'])
#df
```

```python
# Instead of dropping the Name column, we can simply remove it to FirstName
df = df.rename(columns={'Name':'FirstName'})

# And now on to splitting the LongName column
df['FirstName'] = df['LongName'].apply(lambda x: x.split()[0])
df['LongName'] = df['LongName'].apply(lambda x: " ".join(x.split()[1:]))  # Remove the first name in LongName

df = df.rename(columns={'LongName':'LastName'})
```

1. We found all the abbreviated name exceptions
2. We changed the first name in LongName wherever possible
3. Finally, we changed the name of the Name column to FirstName, took it and removed it from LongName, and changed the name of this column to LastName
